```python
import pandas as pd
import os

df = pd.read_csv("mercedes_benz_sales_2020_2025.csv")

df.to_parquet("mercedes_benz_sales_2020_2025.parquet", 
              engine="pyarrow", 
              compression="snappy", 
              index=False)

# 용량 확인
csv_mb     = os.path.getsize("mercedes_benz_sales_2020_2025.csv")     / 1024**2
parquet_mb = os.path.getsize("mercedes_benz_sales_2020_2025.parquet") / 1024**2
print(f"CSV:     {csv_mb:.1f} MB")
print(f"Parquet: {parquet_mb:.1f} MB")
print(f"압축률:  {parquet_mb/csv_mb*100:.1f}%")
```

```python
import pandas as pd
import os

df = pd.read_csv("mercedes_benz_sales_2020_2025.csv")

# 반복 문자열 컬럼 category 변환 + zstd 압축
cat_cols = ["Model", "Region", "Color", "Fuel Type", "Turbo"]
df[cat_cols] = df[cat_cols].astype("category")

df.to_parquet("mercedes_benz_final.parquet", 
              engine="pyarrow", 
              compression="zstd", 
              index=False)

mb = os.path.getsize("mercedes_benz_final.parquet") / 1024**2
print(f"최종 크기: {mb:.1f} MB")
```
